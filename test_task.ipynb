## **Задача:**  
Разработайте алгоритм автоматической разметки временных интервалов, во время которых респондент находился в **возбужденном** и **нейтральном** эмоциональном состоянии исходя из данных о размере зрачка. Работаем в предположении, что размер зрачка является индикатором эмоционального состояния испытуемого: увеличение размера зрачка свидетельствует об эмоциональном возбуждении.

### Описание тестирования:

Тестирование состоит из последовательности этапов, имитирующих различные условия взаимодействия респондента с торговым пространством. Каждый этап моделирует определенное когнитивное и эмоциональное состояние, позволяя анализировать изменения физиологических показателей, таких как размер зрачка.  

- **Baseline** – этап характеризуется отсутствием внешних стимулов. Испытуемый наблюдает пустой стеллаж без товаров, что создает контрольное состояние с минимальной когнитивной и эмоциональной нагрузкой.  
- **Neutral 1** – этап взаимодействия с нейтральным торговым пространством, в котором отсутствуют привлекательные товарные категории и ценовые индикаторы. Данный этап не предполагает вовлеченности в процесс принятия потребительских решений.  
- **Negative** – этап, связанный с воздействием отрицательных стимулов, включая завышенные цены и неряшливый внешний вид торгового пространства (например, грязь на полках). Предполагается, что данные факторы могут вызывать отрицательную эмоциональную реакцию.  
- **Neutral 2** – повторение нейтрального этапа для контроля возможных изменений в состоянии испытуемого после воздействия негативного стимула.  
- **Positive** – этап, предполагающий воздействие положительных стимулов, таких как наличие привлекательных товарных категорий и сниженные цены. Данный стимул может вызывать положительную эмоциональную реакцию.  

Этапы тестирования организованы в фиксированной последовательности, что позволяет контролировать влияние стимулов и анализировать динамику изменения когнитивных и эмоциональных состояний испытуемых.
